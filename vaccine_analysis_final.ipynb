```python
#Define a function to normalize VALUE and set POP_TYPE

def normalize_value(row):
    denominator = row['DENOMINATOR']  # Use 'DENOMINATOR' instead of 'UNIT'
    value = row['INCIDENCE_RATE']     # Use 'INCIDENCE_RATE' instead of 'VALUE'

    if pd.isna(value):
        return value, None  # Keep missing values as they are

    # Normalize to "per 100,000"
    if denominator == 'per 1,000,000 total population':
        return value / 10, 'total_population'
    elif denominator == 'per 1,000,000 <15 population':
        return value / 10, 'under15_population'
    elif denominator == 'per 1,000 live births':
        return value * 100, 'live_births'
    elif denominator == 'per 10,000 live births':
        return value * 10, 'live_births'
    elif denominator == 'per 1000 live births':
        return value * 100, 'live_births'
    else:
        return value, 'unknown'

# Apply the normalization function to the full DataFrame
df_incidence[['INCIDENCE_RATE_NORM', 'POP_TYPE']] = df_incidence.apply(normalize_value, axis=1, result_type='expand')

# Add a new column for the normalized unit
df_incidence['DENOMINATOR_NORM'] = 'per 100,000'
```

```python
import pandas as pd
import mysql.connector

# MySQL connection details
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'prijags20',
    'database': 'vaccination_db'
}

# Connect to MySQL
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor()

# Function to insert data into MySQL table from DataFrame
def insert_data(df, table_name):
     # Escape reserved words in column names by wrapping them with backticks
    cols = ', '.join([f"`{col}`" for col in df.columns])
    placeholders = ', '.join(['%s'] * len(df.columns))
    sql = f"INSERT INTO {table_name} ({cols}) VALUES ({placeholders})"
    for i, row in df.iterrows():
        cursor.execute(sql, tuple(row))
    conn.commit()
    print(f"Inserted {len(df)} rows into {table_name}")

```
